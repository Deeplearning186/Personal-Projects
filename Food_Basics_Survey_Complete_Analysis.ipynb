# Food Basics Survey Data Analysis

This notebook contains the data analysis and visualization for the Food Basics Survey. The analysis focuses on understanding customer satisfaction, demographic differences, and shopping behaviors.
    

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
from google.colab import files
uploaded = files.upload()

import io
data = pd.read_excel(io.BytesIO(uploaded['Food_Basics_Survey_Cleaned.xlsx']))

```


## Distribution of Gender, Age Range, and Overall Satisfaction Ratings

Let's start by looking at the distribution of respondents by gender, their age range, and their overall satisfaction ratings.
    

```python

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Function to create count plots for categorical columns
def plot_categorical_distribution(column_name, title, rotation=0):
    plt.figure(figsize=(10, 6))
    sns.countplot(y=column_name, data=data, order=data[column_name].value_counts().index)
    plt.title(title)
    plt.xticks(rotation=rotation)
    plt.show()

# Plotting the distribution of 'Gender', 'Age Range', and 'Overall Satisfaction'
plot_categorical_distribution('Gender', 'Distribution of Gender')
plot_categorical_distribution('Age Range', 'Distribution of Age Range')
plot_categorical_distribution('Overall Satisfaction', 'Distribution of Overall Satisfaction Ratings')
```


## Analysis of Satisfaction Ratings by Demographic Groups

Next, we analyze how satisfaction ratings vary across different demographic groups, focusing on gender and age range.
    

```python

# Function to create box plots for examining the relationships between satisfaction metrics and demographic groups
def plot_demographic_differences(column_name, demographic, title):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=demographic, y=column_name, data=data, palette="Set2")
    plt.title(title)
    plt.show()

# Analyzing satisfaction ratings by demographic groups
plot_demographic_differences('Product Availability Rating', 'Gender', 'Product Availability Rating by Gender')
plot_demographic_differences('Fresh Produce Quality Rating', 'Gender', 'Fresh Produce Quality Rating by Gender')
plot_demographic_differences('Checkout Efficiency Rating', 'Gender', 'Checkout Efficiency Rating by Gender')
```


## Shopping Frequency and Influence of Promotions

We'll also explore how shopping frequency and the influence of promotions vary among different demographic groups.
    

```python

# Plotting the distribution of 'Shopping Frequency' by 'Gender' and 'Age Range'
plt.figure(figsize=(10, 6))
sns.countplot(y='Shopping Frequency', hue='Gender', data=data, palette="Paired")
plt.title('Shopping Frequency by Gender')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Shopping Frequency', hue='Age Range', data=data, palette="Paired")
plt.title('Shopping Frequency by Age Range')
plt.show()

# Analyzing the influence of promotions by 'Gender' and 'Age Range'
plt.figure(figsize=(10, 6))
sns.countplot(y='Influence of Promotions', hue='Gender', data=data, palette="Paired")
plt.title('Influence of Promotions by Gender')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Influence of Promotions', hue='Age Range', data=data, palette="Paired")
plt.title('Influence of Promotions by Age Range')
plt.show()
```
